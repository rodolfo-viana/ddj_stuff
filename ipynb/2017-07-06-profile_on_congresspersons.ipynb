### Introdução

A presente análise tem por objetivo traçar um retrato dos deputados federais eleitos em 2014 por meio do gastos ressarcidos pela Cota para o Exercício da Atividade Parlamentar (CEAP), foco do trabalho da **Operação Serenata de Amor**, desde o começo da atual legislatura.

Os dados utilizados no estudo compreendem o período de janeiro de 2015 à presente data. Eles foram coletados a partir do sistema de dados abertos da Câmara dos Deputados<sup>1</sup>. A análise desses dados é realizada com o uso de Python.

São focos deste estudo:

* a quantidade total de reembolsos solicitados;
* o valor total dos reembolsos;
* o valor médio por deputado no período da atual legislatura;
* o valor médio por deputado ao mês;
* a proporção de cada categorias de despesas no valor total;
* a relação entre Estados e categorias de despesas; e
* o valor médio por deputado, por Estado.

Nomes de deputados e partidos políticos não fazem parte da análise<sup>2</sup>.

Após os textos dos resultados encontrados, são apresentados os códigos usados na execução da análise, com anotações descritivas nos campos. Isso reforça o detalhamento do trabalho aqui realizado.

<sup>1</sup> Como a Câmara fez a migração dos dados abertos para a versão 2.0 recentemente, a Serenata de Amor está empenhada em atualizar suas ferramentas para contemplas mudanças na nova versão. Desta forma, excepcionalmente para este estudo, serão utilizados os arquivos `.xml` -- cuja atualização é mais frequente -- convertidos em `.csv`, e não o conjunto de dados baixados pela Serenata de Amor.

<sup>2</sup> O autor do estudo acredita que utilizar nomes e siglas poderia partidarizar a leitura dos resultados e, portanto, tais dados são descartados.

Considerando o número de beneficiários da CEAP no período, a média de reembolso per capita é de 596.735,17 desde janeiro de 2015. Por mês, 19.891,17 por deputado. 

Este número, contudo, não retrata a realidade: há pessoas que ocuparam o posto de deputado por pouco tempo, enquanto outras estão na Câmara desde o começo da legislatura. A certeza é que o número de deputados em atuação ao mesmo tempo nunca superou 513 parlamentares. 

Com isso, a média total per capita sobe para 965.477,96 reais no período, ou 32.182,59 reais mensais por deputado.

### Linhas de código

```python
# Importação das ferramentas de análise
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# Seleção dos dados que serão utilizados na análise
dataset2017 = pd.read_csv('../data/Ano-2017.csv', sep=';',
                         dtype={'idecadastro': np.str,
                                'ideDocumento': np.str},
                         converters={'vlrLiquido': lambda x: float(x.replace(',','.'))},
                         low_memory=False)
dataset2016 = pd.read_csv('../data/Ano-2016.csv', sep=';',
                         dtype={'idecadastro': np.str,
                                'ideDocumento': np.str},
                         converters={'vlrLiquido': lambda x: float(x.replace(',','.'))},
                         low_memory=False)
dataset2015 = pd.read_csv('../data/Ano-2015.csv', sep=';',
                         dtype={'idecadastro': np.str,
                                'ideDocumento': np.str},
                         converters={'vlrLiquido': lambda x: float(x.replace(',','.'))},
                         low_memory=False)

# Organização dos dados, limpeza de colunas descartáveis, renomeação de categorias
dataset = pd.concat([dataset2017, dataset2016, dataset2015])
dataset = dataset[['txNomeParlamentar', 
                   'idecadastro', 
                   'sgUF', 
                   'numSubCota', 
                   'txtDescricao', 
                   'datEmissao', 
                   'vlrLiquido', 
                   'numMes', 
                   'numAno', 
                   'ideDocumento']]
dataset['txtDescricao'] = dataset['txtDescricao'].astype('category')
dataset['txtDescricao'].cat.rename_categories(['Assinatura de publicações', 
                                              'Combustíveis e lubrificantes',
                                              'Consultorias, pesquisas e trabalhos técnicos',
                                              'Divulgação de atividade parlamentar', 
                                              'Emissão de bilhete aéreo',
                                              'Alimentação do parlamentar',
                                              'Hospedagem fora do DF',
                                              'Aluguel de aeronaves',
                                              'Aluguel de embarcações',
                                              'Aluguel de carros', 
                                              'Manutenção de escritório',
                                              'Cursos, palestras ou eventos', 
                                              'Passagens aéreas',
                                              'Passagens terrestres ou marítimas',
                                              'Segurança particular',
                                              'Táxi, pedágio e estacionamento', 
                                              'Serviços postais', 
                                              'Telefonia'], inplace=True)

# Contagem de notas fiscais em análise, por categoria
print("Quantidade de notas fiscais em análise, por categoria:")
dataset['txtDescricao'].value_counts()
```

```python
# Soma de notas fiscais em análise
print("Quantidade de notas fiscais em análise:", dataset.ideDocumento.count())

# Soma dos valores
print("Valor total de reembolsos: R$", dataset.vlrLiquido.sum())
```

```python
# Formatação dos valores em milhares de reais (para os gráficos a seguir)
columns = ['vlrLiquido',]
for column in columns:
    value_in_thousands = []
    for item, row in dataset.iterrows():
        value_row = row[column]
        value_in_thousands.append(value_row/10**3)
    name_variable = '{}_mil'.format(column)
    variable = pd.Series(value_in_thousands)
    dataset[name_variable] = variable.values
```

```python
# Soma de pessoas que utilizaram a CEAP na atual legislatura
print("Pessoas que utilizaram a CEAP:", dataset.idecadastro.nunique())
```

```python
# Divisão de valores de reembolso por beneficiários para média geral
print("Média de reembolso por usuário da CEAP: R$", dataset.vlrLiquido.sum() / \
                                                    dataset.idecadastro.nunique())

# Divisão de valores de reembolso por beneficiários e pelos meses, para média mensal
print("Média de reembolso por usuário da CEAP por mês: R$", (dataset.vlrLiquido.sum() / \
                                                             dataset.idecadastro.nunique()) / 30)

# Divisão de valores de reembolso por 513 deputados para média geral
print("Média de reembolso por deputado: R$", dataset.vlrLiquido.sum() / 513)

# Divisão de valores de reembolso por 513 deputados e pelos meses, para média mensal
print("Média de reembolso por deputado por mês: R$", (dataset.vlrLiquido.sum()) / 513 / 30)
```

```python
# Soma de valores por categoria
dataset_per_subquota = dataset.groupby(['txtDescricao'])\
                                       .agg({'vlrLiquido': sum})\
                                       .sort_values(by='vlrLiquido', ascending=False)\
                                       .reset_index()

# Extração do percentual
dataset_per_subquota['porcentagem'] = (dataset_per_subquota['vlrLiquido']\
                                      / dataset_per_subquota['vlrLiquido'].sum()) * 100

# Formatação do decimal para dois dígitos
dataset_per_subquota['vlrLiquido'] = dataset_per_subquota['vlrLiquido'].map("{0:.2f}".format)
dataset_per_subquota['porcentagem'] = dataset_per_subquota['porcentagem'].map("{0:.2f}".format)

# Resultado
dataset_per_subquota
```

```python
# Começando a agregar despesas por categoria e Estado...
keys = ['txtDescricao', 'idecadastro', 'sgUF']
br_per_capita_per_category = dataset.groupby(keys)['vlrLiquido']\
                             .agg({'soma': np.sum, 'gastos': len})
br_per_capita_per_category = br_per_capita_per_category.reset_index()

subquota_list = list(set(br_per_capita_per_category['txtDescricao']))
state_list = list(set(br_per_capita_per_category['sgUF']))
values_list = list(set(br_per_capita_per_category['soma']))
```

```python
# Por categoria...
for subquota in subquota_list:
    print(subquota, br_per_capita_per_category\
          .query('txtDescricao == "{}"'\
          .format(subquota))\
          .shape)
```

```python
# Soma dos valores por Estado
dataset_per_state = dataset.groupby(['sgUF'])\
                                    .agg({'vlrLiquido':sum})\
                                    .sort_values(by='vlrLiquido', ascending=False)\
                                    .reset_index()

# Formatação do decimal para dois dígitos
dataset_per_state['vlrLiquido_legivel'] = dataset_per_state['vlrLiquido'].map("{0:.2f}".format)

# Resultado (mantendo a notação científica para uso posterior)
dataset_per_state
```

```python
# Criação de coluna com quantidade de assentos por Estado
dataset_per_state['assentos'] = [70, 53, 46, 39, 31, 30, 25, 22, 18, 17, 16, 17, 
                                 12, 10, 8, 8, 8, 10, 9, 8, 8, 8, 8, 8, 8, 8, 8]

# Definição de média per capita
dataset_per_state['media_per_capita'] = dataset_per_state['vlrLiquido'] / dataset_per_state['assentos']
dataset_per_state['media_per_capita_legivel'] = dataset_per_state['media_per_capita'].map("{0:.2f}".format)

# Resultado (mantendo a notação científica)
dataset_per_state
```

```python
# Média per capita em gráfico
sns.set(font_scale=1.2)
plt.figure(figsize=(16, 8))
g = sns.barplot(x='sgUF',
                y='media_per_capita',
                data=dataset_per_state)
g.axes.set_title('CEAP: Média per capita por Estado')
g.set_ylabel('')    
g.set_xlabel('')
sns.despine()
```

```python
# Observação de valor médio ('mean') e desvio-padrão ('std') para futuros gráficos
dataset_per_state.media_per_capita.describe().map("{0:.2f}".format)
```

```python
# Soma de valores (em milhares de reais) por categoria e Estado
dataset_per_subquota_per_state = dataset.groupby(['txtDescricao', 'sgUF'])\
                                                 .agg({'vlrLiquido_mil':sum})\
                                                 .sort_values(by='vlrLiquido_mil', ascending=False)\
                                                 .reset_index()

# Giro de eixo dos dados para usar no mapa de calor
dataset_per_subquota_per_state_pivot = dataset_per_subquota_per_state\
                                       .pivot(index='sgUF',\
                                              columns='txtDescricao',\
                                              values='vlrLiquido_mil')

# Criação de mapa de calor
sns.set(font_scale=1.2)
plt.figure(figsize=(18, 10))
ax = sns.heatmap(dataset_per_subquota_per_state_pivot, \
                 cmap='magma_r', \
                 annot=True, \
                 fmt='.0f', \
                 cbar=True, \
                 vmax=18000, \
                 annot_kws={"size":12})
ax.set_title('CEAP: Gastos por Estado e categoria de despesa (valores em R$ mil)')
ax.set_ylabel('')    
ax.set_xlabel('')
sns.despine()
```
