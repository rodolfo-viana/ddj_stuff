```python
# Checagem com resultado do Metrô
bike_limpo['FE_PESS'].sum()
```

```python
# Segregação por sexo
bike_sexo = bike_limpo.groupby('SEXO').agg({'FE_PESS': 'sum'}).reset_index()
bike_sexo['QUANTIDADE'] = round(bike_sexo['FE_PESS']).astype(int)
bike_sexo['PROP'] = (bike_sexo['FE_PESS'] / bike_sexo['FE_PESS'].sum()) * 100
bike_sexo
```

```python
# Segregação por faixa etária
fx_et = []
for row in bike_limpo['IDADE']:
    if row < 18:
        fx_et.append('Menor de 18 anos')
    elif row >= 18 and row < 25:
        fx_et.append('18 a 24 anos')
    elif row >= 25 and row < 31:
        fx_et.append('25 a 30 anos')
    elif row >= 31 and row < 41:
        fx_et.append('31 a 40 anos')
    elif row >= 41 and row < 51:
        fx_et.append('41 a 50 anos')
    elif row >= 51 and row < 61:
        fx_et.append('51 a 60 anos')
    else:
        fx_et.append('61 anos ou mais')

bike_limpo['FAIXA ETÁRIA'] = fx_et
bike_fx_et = bike_limpo.groupby('FAIXA ETÁRIA').agg({'FE_PESS': 'sum'}).reset_index()
bike_fx_et['QUANTIDADE'] = round(bike_fx_et['FE_PESS']).astype(int)
bike_fx_et['PROP'] = (bike_fx_et['FE_PESS'] / bike_fx_et['FE_PESS'].sum()) * 100
bike_fx_et
```

```python
# Segregação por posse de carro ou moto
cr_mt = []
bike_limpo['CARRO E/OU MOTO'] = bike_limpo['QT_AUTO'] + bike_limpo['QT_MOTO']
for row in bike_limpo['CARRO E/OU MOTO']:
    if row == 0:
        cr_mt.append('Não')
    elif row > 0:
        cr_mt.append('Sim')
    else:
        cr_mt.append('Não respondeu')
bike_limpo['CARRO_MOTO'] = cr_mt
bike_cr_mt = bike_limpo.groupby('CARRO_MOTO').agg({'FE_PESS': 'sum'}).reset_index()
bike_cr_mt['QUANTIDADE'] = round(bike_cr_mt['FE_PESS']).astype(int)
bike_cr_mt['PROP'] = (bike_cr_mt['FE_PESS'] / bike_cr_mt['FE_PESS'].sum()) * 100
bike_cr_mt
```
